## **Handling missing values**

### **Missing values in the categorical features**

```python
categorical_features_with_NaN = [feature for feature in X_train.columns if X_train[feature].isnull().sum()>=1 and X_train[feature].dtype == 'O']

for feature in categorical_features_with_NaN:
  percent_nan = (X_train[feature].isnull().mean())*100
  print(f'{feature} has {np.round(percent_nan, decimals=4)} % NaN values.')
```

##### **Observations:**
1. Some features like `Alley`, `FireplaceQu`, `PoolQC`, `Fence`, `MiscFeature` have an overwhelmingly huge amount of missing values. At first glance, it is tempting to drop these features. But, we should look at the "data_description.txt" file first. In it, we can see that "NA" is actually a category in `Alley` which stands for "No alley access". Hence, we can just replace the missing values here with a string "Missing".
2. The same thing as `Alley` is observed in `FireplaceQu`, `PoolQC`, `Fence`, and `MiscFeature`. Hence, we will replace all the missing values in these variables with the string "Missing", and add a missing indicator to capture this "missingness".
3. For the categorical variables containing $<5\%$ missing values, we will do mode or most frequent value imputation, whereas for the categorical variables containing $>5\%$ missing values, we will to missing value imputation as explained in the second point above.

#### **Mode imputation for $<5\%$ missing values**

```python
less_than_five_pc = []
for feature in categorical_features_with_NaN:
  percent_nan = (X_train[feature].isnull().mean())*100
  if percent_nan<5:
    less_than_five_pc.append(feature)

less_than_five_pc
```

```python
from feature_engine.imputation import CategoricalImputer
```

```python
imputer_frequent_less_five = CategoricalImputer(
    imputation_method='frequent',
    variables=less_than_five_pc
)

imputer_frequent_less_five.fit(X_train)
```

```python
X_train = imputer_frequent_less_five.transform(X_train)
X_test = imputer_frequent_less_five.transform(X_test)
```

#### **'Missing' imputation for $>5\%$ missing values**

##### **Adding a missing indicator**

```python
more_than_five_pc = [feature for feature in categorical_features_with_NaN if feature not in less_than_five_pc]
more_than_five_pc
```

```python
from feature_engine.imputation import AddMissingIndicator
missing_indicator_more_five = AddMissingIndicator(variables=more_than_five_pc)

missing_indicator_more_five.fit(X_train)
```

```python
X_train = missing_indicator_more_five.transform(X_train)
X_test = missing_indicator_more_five.transform(X_test)
```

##### **Imputing 'missing'**

```python
missing_imputer = CategoricalImputer(imputation_method='missing',
                                     variables=more_than_five_pc)
missing_imputer.fit(X_train)
```

```python
X_train = missing_imputer.transform(X_train)
X_test = missing_imputer.transform(X_test)
```

## **Handling Rare Labels in Categorical Features**

We will rename the labels as "Rare" for labels present in less than $5\%$ of the observations in a particular categorical feature. We define this $5\%$ as the *tolerance*. We will do this for each categorical feature.

```python
categorical_features = [feature for feature in X_train.columns if X_train[feature].dtype == 'O']
print(f'There are total {len(categorical_features)} categorical variables. They are:')
categorical_features
```

```python
def find_non_rare_labels(df, variable, tolerance):
    
    temp = df.groupby([variable])[variable].count() / len(df)
    
    non_rare = [x for x in temp.loc[temp>tolerance].index.values]
    
    return non_rare

def rare_encoding(X_train, X_test, variable, tolerance):

    X_train = X_train.copy()
    X_test = X_test.copy()

    # find the most frequent category
    frequent_cat = find_non_rare_labels(X_train, variable, tolerance)

    # re-group rare labels
    X_train[variable] = np.where(X_train[variable].isin(
        frequent_cat), X_train[variable], 'Rare')
    
    X_test[variable] = np.where(X_test[variable].isin(
        frequent_cat), X_test[variable], 'Rare')

    return X_train, X_test
```

```python
for feature in categorical_features:
  X_train, X_test = rare_encoding(X_train, X_test, feature, 0.05)
```

```python
X_train[categorical_features].head(10)
```

We can see that we have replaced the labels appearing $<5\%$ of the time with the label "Rare".

## **Ordinal Encoding Categorical Features**
Now, we know that our dataset already has a lot of features. Carrying out one-hot encoding will increase the feature space even more. Hence, we are choosing to do ordinal categorical encoding. This will also be helpful to get a monotonic relationship between the encoded labels and the target `SalePrice`, which may improve the model performance. This is known as *target guided encoding*.

```python
from feature_engine.encoding import OrdinalEncoder
ordinal_enc = OrdinalEncoder(
    encoding_method='ordered',
    variables=categorical_features
)
```

```python
ordinal_enc.fit(X_train, y_train)
```

```python
X_train = ordinal_enc.transform(X_train)
X_test = ordinal_enc.transform(X_test)
```

## **Feature Scaling**
Feature scaling is a very important step. We will use the MinMaxScaler to scale the values of all the variables such that they lie in the interval $[0,1]$.

```python
features_to_feature_scale = [feature for feature in X_train.columns if feature not in ['Id', 'SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train[features_to_feature_scale])
```
